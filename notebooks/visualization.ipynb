## Visualization for the Cleaned Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#reading the cleaned imdb file from the data folder

df=pd.read_csv('../data/imdb-cleaned.csv')
df
```

## Univariate Analysis 

### IMDB Rating Histogram

- The graph is the **Right Skewed Graph** as the IMDB rating increases the number of the films get decreases
- Rating of **7 to 7.9** contain the maximum number of the movies.
- **KEY INSIGHTS :** HIGh IMDB rating contain less number of movies

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='IMDB',color='violet')
plt.title('IMDB Ratings Analysis')
plt.xlabel('IMDB Ratings')
plt.ylabel('Total Number of Films')
plt.grid()
plt.show()
```

```python
#KDE Plot for the same analysis

plt.figure(figsize=(10,6))
sns.kdeplot(data=df,x='IMDB',color='green',linewidth=2.5)
plt.title('IMDB Ratings Analysis')
plt.xlabel('IMDB Ratings')
plt.ylabel('Total Number of Films')
plt.grid()
plt.show()
```

### Gross Per Million Analysis

- The graph is the **RIGHT SKEWED GRAPH** only few number of films had earned more than 800M $

- 400+ films have earned less than 100M $
- Only 2 to 3 films have earned more than 800M+ $

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Gross_per_M',color='violet')

plt.xlabel('Gross Per Million')
plt.ylabel('Total Number of Films')

plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=df,x='Gross_per_M',color='green',linewidth=3)

plt.xlabel('Gross Per Million')
plt.ylabel('Total Number of Films')

plt.grid()
plt.show()
```

### Duration Analysis

- Average Duration of the film is around **90 to 120 minutes**
- the graph is the **RIGHT SKEWED GRAPH** the number of film decreases as the duration increases
- ONLY 1 to 2 films have duration higher than 300+ or less than 50 minutes

- **KEY INSIGHTS :** HIGH DURATION contains LOW NUMBER OF FILMS

```python
plt.figure(figsize=(10,6))

sns.histplot(data=df,x='Duration',color='violet')
plt.xlabel('Total Durations')
plt.ylabel('Number of Films')
plt.title('Duartion Analysis')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.kdeplot(data=df,x='Duration',color='green',linewidth=3)
plt.xlabel('Total Durations')
plt.ylabel('Number of Films')
plt.title('Duartion Analysis')
plt.grid()
plt.show()
```

### Metascore Analysis

- The Graph is the **LEFT SKEWED GRAPH** only few films contain the less number of metascore
- More than 200+ films have metascore of 79
- Less than 50 films have metascore of 100

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df,x='Metascore',color='violet')

plt.xlabel('Metascore of Movies')
plt.ylabel('Number of Films')
plt.title('Metascore Analysis')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=df,x='Metascore',color='green',linewidth=3)

plt.xlabel('Metascore of Movies')
plt.ylabel('Number of Films')
plt.title('Metascore Analysis')
plt.grid()
plt.show()
```

### Genre Frequency

- **DRAMA** is the most popular genre of all time total 289+ movies
- Less favourites are **Fantasy and Thriller** less than 5 movies


```python
plt.figure(figsize=(10,6))
#here we need to calculate the total number of genre

temp=df.groupby('Genres')['Name'].count().sort_values(ascending=False)
sns.barplot(x=temp.index,y=temp.values,hue=temp.index)
plt.xlabel('Genres')
plt.ylabel('Number of Films')
plt.title('Genres Frequency Analysis')
plt.xticks(rotation=45)

plt.grid()
plt.show()
```

### Decade Analysis

- the number of film increases as the Decade increases
- **2010** have the highest number of the films
- **1920** have the lowest number of the films
- **2020** is currently evaluating and in progress 

```python
plt.figure(figsize=(10,6))
#here we need to calculate the total number of genre

dec=df.groupby('Decade')['Name'].count().sort_values(ascending=False)
sns.barplot(x=dec.index,y=dec.values,hue=dec.index,palette='deep')
plt.xlabel('Decades')
plt.ylabel('Number of Films')
plt.title('Decade Frequency Analysis')


plt.grid()
plt.show()
```

## Bivariate Analysis



### IMDB VS GROSS Analysis

- **Negative Correlation** as the IMDB Rating increases the Earning of the film decreases
- The film less than IMDB of **7 to 8.3** have earned more than 800M+
- The highest rated film have less than 50M earning


```python
plt.figure(figsize=(10,6))

sns.regplot(data=df,x='IMDB',y='Gross_per_M',scatter_kws={'color':'darkblue','s':50,'alpha':0.4},line_kws={'color':'red','linewidth':3})
plt.xlabel('IMDB Ratings')
plt.ylabel('Gross Per Million')
plt.title('IMDB vs Gross Analysis')
plt.grid()
plt.show()
```

### IMDB VS Metascore Analysis

- This graph is used to show the **Positive Correlation**
- Film with the HIGH IMDB Rating have HIGH Metascore 
- There are also some films which have high IMDB rating but Metascore is low

```python
plt.figure(figsize=(10,6))
sns.regplot(data=df,x='IMDB',y='Metascore',
            scatter_kws={'color':'green','s':50,'alpha':0.4},
            line_kws={'color':'red','linewidth':3}
        )

plt.xlabel('IMDB Ratings')
plt.ylabel('Metascore')
plt.title('IMDB VS Metascore Analysis')
plt.grid()
plt.show()
```

### IMDB VS Duration 

- There are less number of films whose **IMDB Rating** is 8.75 to 9.5 and there duration is less
- the highest number of films have duration of about **60 to 120 hours**
- One film with IMDB=8.1 have highest duration of more than 300+ minutes

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='IMDB',y='Duration',color='black',alpha=0.6,s=90)
plt.xlabel('IMDB Rating')
plt.ylabel('Film Duration')
plt.title('IMDB VS Duration Analysis')
plt.grid()
plt.show()
```

### Votes Per K Vs Gross Per M

- In the given graph the film which have lower votes have high earning
- The film which have most number of votes have low earning then other films
- Many film with about **100 votes** have **800M+ $ earning**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='Votes_per_K',y='Gross_per_M',s=90,color='blue',alpha=0.6)
plt.xlabel('Votes Per K')
plt.ylabel('Gross Per M')
plt.title('Votes Vs Gross Analysis')
plt.grid()
plt.show()
```
